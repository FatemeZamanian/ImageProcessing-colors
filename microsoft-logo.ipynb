```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
r=np.zeros((720,1440),dtype='uint8')
g=np.zeros((720,1440),dtype='uint8')
b=np.zeros((720,1440),dtype='uint8')

r[240:360,240:360]=255
g[240:360,240:360]=80
b[240:360,240:360]=34

r[240:360,370:490]=127
g[240:360,370:490]=220
b[240:360,370:490]=8

r[370:490,240:360]=28
g[370:490,240:360]=109
b[370:490,240:360]=209

r[370:490,370:490]=255
g[370:490,370:490]=200
b[370:490,370:490]=3


res=cv2.merge((r,g,b))


plt.imshow(res)


```

```python
res=cv2.putText(res,'Microsoft',(520,420),cv2.FONT_HERSHEY_DUPLEX,4.5,(255,255,255),12,cv2.LINE_AA)
plt.imshow(res)
```

```python
row,col,ch=res.shape
for i in range(row):
    for j in range(col):
        if res[i,j][0]==0 and res[i,j][1]==0 and res[i,j][2]==0:
            res[i,j]=90
plt.imshow(res)
```
